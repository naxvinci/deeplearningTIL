```python
## RNN에 의한 일본경제 평균 주가(종가) 예측
## 과거 30일분의 주가로부터 당일의 주가를 예측
## 과거 300~61일분을 훈련용 데이터
## 과거 60~31일분을 검증용 데이터
## 과거 30~0일분을 테스트용 데이터

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# from tensorflow.keras.models import Sequential, Input
# from tensorflow.keras.layers import Dense, LSTM, GRU
# from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, LSTM, GRU, Input
from tensorflow.keras.optimizers import RMSprop
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

## https://indexes.nikkei.co.jp/nkave/historical/nikkei_stock_average_daily_jp.csv

data_file = 'nikkei_stock_average_daily_jp.csv'

## 당일 주가를 예측하기 위해 필요한 과거 일수
lookback = 30

epochs = 10

## 데이터파일 읽기
## 데이터 일자를 index 설정
## 마지막 행은 데이터가 없기 때문에, 스킵
df = pd.read_csv(data_file, index_col=0)

## 종가
closing_price = df[['종가']].values
```

```python
## 훈련, 검증, 테스트용 데이터 작성
## 과거 30일분의 주가로부터 당의 주가 계산
def data_split(data, start, end, lookback):
    length = abs(start-end)
    
    X = np.zeros((length, lookback))
    y = np.zeros((length, 1))
    
    for i in range(length):
        j = start - lookback + i
        k = j + lookback
        
        X[i] = data[j:k, 0]
        y[i] = data[k, 0]
        
    return X, y
```

```python
## 훈련, 검증, 테스트용 데이터
(X_train, y_train) = data_split(closing_price, -300, -60, lookback)
(X_valid, y_valid) = data_split(closing_price, -60, -30, lookback)
(X_test, y_test) = data_split(closing_price, -30, 0, lookback)
```

```python
plt.title('Changes in Nikkei Stock Average (closing price)')
plt.plot(range(len(closing_price)), closing_price)
plt.show()
```

```python
## 표준화
## X만 차원을 변환 (2차원 -> 3차원)
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
X_train_std = scaler.transform(X_train).reshape(-1, lookback, 1)
X_valid_std = scaler.transform(X_valid).reshape(-1, lookback, 1)
X_test_std = scaler.transform(X_test).reshape(-1, lookback, 1)
```

```python
scaler.fit(y_train)
y_train_std = scaler.transform(y_train)
y_valid_std = scaler.transform(y_valid)
```

```python
X_train_std.shape[-1]
```

```python
result = model.fit(X_train_std, y_train_std, 
                   verbose=1,
                   epochs=epochs, 
                   batch_size=64, 
                   shuffle=True, 
                   validation_data=(X_valid_std, y_valid_std))
```

```python
## 訓練の損失値をプロット
epochs = range(len(result.history['loss']))
plt.title('Loss')
plt.plot(epochs, result.history['loss'], 'bo', alpha=0.6, marker='.', label='Train', linewidth=1)
plt.plot(epochs, result.history['val_loss'], 'r', alpha=0.6, label='Validate', linewidth=1)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
## 予測値
df_predict_std =  pd.DataFrame(model.predict(X_test_std), columns=['予測値'])

## 予測値を元に戻す
predict = scaler.inverse_transform(df_predict_std['予測値'].values)

## 予測結果をプロット
pre_date = df.index[-len(y_test):].values
plt.title('Target close price and predict price')
plt.plot(pre_date, y_test, 'b', alpha=0.6, marker='.', label='Target', linewidth=1)
plt.plot(pre_date, predict, 'r', alpha=0.6, marker='.', label='Train', linewidth=1)
plt.xticks(rotation=70)
plt.legend()
plt.grid(True)
plt.show()
```

```python
## RMSE 계산 (평균제곱근의 편차)
print('RMSE: %.3f' %  
       np.sqrt(mean_squared_error(y_test, predict)))
```
