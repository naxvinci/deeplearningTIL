```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from keras.layers import *
from keras.models import *
from keras.utils import *
from sklearn.preprocessing import *
```

```python
label = ['feat_{}'.format(i) for i in range(73)]
df = pd.read_csv('eighthr.data', names=label)
df
```

```python
df.dtypes # 숫자이어야 하는데 object(string) -> 숫자 변환해야 함
```

```python
df = df.apply(pd.to_numeric, errors='coerce') # 숫자를 NaN으로 (오류일 경우)
df
```

```python
df.dtypes # 숫자이어야 하는데 object(string) -> 숫자 변환해야 함
```

```python
# NaN 제거해야 함 
df.dropna(inplace=True) # 바로 적용
```

```python
df
```

```python
df.describe()
```

```python
# 마지막 label이 정답 (feat_72)
```

```python
Y = df['feat_72']
```

```python
Y
```

```python
Y = to_categorical(Y)
```

```python
Y
```

```python
df.drop(["feat_72"], axis=1, inplace=True)
```

```python
df
```

```python
X_train = np.asarray(df[:-100].values.tolist(), dtype=np.float64)
X_test = np.asarray(df[-100:].values.tolist(), dtype=np.float64)

Y_train = Y[:-100]
Y_test = Y[-100:]
```

```python
X_train.shape
```

```python
# (sample, timestamp, feature) 
X_train = X_train.reshape(-1, 10, 72) # 10개로 딱 떨어지지 않기 때문에, 10개로 나울수 있도록 데이터를 자름
```

```python
X_train = X_train[:1700]
Y_train= Y_train[:1700]
```

```python
X_train = X_train.reshape(-1, 10, 72)
```

```python
Y_train = Y_train.reshape(-1, 10, 2)
```

```python
X_train.shape
```

```python
Y_train.shape
```

## Create model

```python
model = Sequential()
model.add(LSTM(128, input_shape=(10, 72), return_sequences=True))
model.add(LSTM(256, return_sequences=True))
model.add(Dense(2, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_train, Y_train, epochs=10, batch_size=1, validation_split=0.1)
```

```python
score = model.evaluate(X_test, Y_test)
print(score)
```
