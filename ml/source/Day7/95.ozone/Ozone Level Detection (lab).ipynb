```python
df.drop(['feat_72'], axis=1, inplace=True) # inplace는 덮어쓰겠다는 의미
```

```python
X_train = np.asarray(df[:-100].values.tolist(), dtype=np.float64)
X_test = np.asarray(df[-100:].values.tolist(), dtype=np.float64)
```

```python
Y_train = Y[:-100]
Y_test = Y[-100:]
```

```python
# (sample, timestep, feature)  -> 10개로 나눠지지 않아
X_train = X_train[:1700]
Y_train = Y_train[:1700]
X_train = X_train.reshape(-1, 10, 72)
Y_train = Y_train.reshape(-1, 10, 2)#  -> shape이 (1747,2)라서 2로 쪼갬
```

```python
print(X_train.shape, Y_train.shape)
```

## Create model

```python
i = Input(shape=(10, 72))
x = LSTM(128, return_sequences=True)(i)
x = LSTM(256, return_sequences=True)(x)
x = Dense(2, activation='softmax')(x) # default는 linear regression

model = Model(i, x)
model.compile(loss='categorical_crossentropy', optimizer='adam',
             metrics=['accuracy'])

model.summary()
```

```python
r = model.fit(X_train, Y_train, epochs=10, batch_size=1, validation_split=0.1) # 비율로 쪼갤때 10퍼센트
```
