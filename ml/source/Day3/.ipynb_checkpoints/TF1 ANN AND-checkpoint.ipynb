```python
import numpy as np
import tensorflow as tf
```

```python
tf.enable_eager_execution()
tf.__version__
```

```python
X = np.array([
    [0., 0.],
    [0., 1.],
    [1., 0.],
    [1., 1.]
], dtype="float32")

y = np.array([
    [0.],
    [0.],
    [0.],
    [1.]
], dtype="float32")
```

```python
X
```

```python
y
```

```python
w = tf.Variable([[1], [2]], dtype="float32")
b = tf.Variable([3], dtype="float32")
```

```python
w
```

```python
b
```

```python
tf.matmul(X, w)
```

```python
tf.matmul(X, w) + b
```

```python
tf.sigmoid(tf.matmul(X, w) + b)
```

```python
optimizer = tf.train.AdamOptimizer(0.01) # 뒤에 숫자는 learning rate다
```

```python
for step in range(5000):
    with tf.GradientTape() as tape:
        hypothesis = tf.sigmoid(tf.matmul(X, w) + b)
        cost = -tf.reduce_mean(y * tf.log(hypothesis) + (1-y) * tf.log(1-hypothesis))
        grads = tape.gradient(cost, [w, b])
    optimizer.apply_gradients(grads_and_vars=zip(grads, [w, b]))
    
    if step % 100 == 0:
        print("=" * 50)
        print("step: {}, cost: {}, w: {}, b: {}".format(step, cost.numpy(), w.numpy(), b.numpy()))
        print("=" * 50)
```

```python
w
```

```python
b
```

```python
predict = tf.sigmoid(tf.matmul(X, w) + b)
predict
```

```python
predict01 = tf.cast(predict > 0.5, dtype="float32")
print("=" * 30)
print("predict01")
print(predict01)
```

```python
ac01 = tf.equal(predict01, y)
ac01
```

```python
ac02 = tf.cast(ac01, dtype="float32")
ac02
```

```python
ac03 = tf.reduce_mean(ac02)
ac03
```

```python
accuracy = tf.reduce_mean(tf.cast(tf.equal(predict01, y), dtype="float32"))
print("=" * 30)
print("accuracy")
print(accuracy.numpy())
```
