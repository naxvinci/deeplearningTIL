```python
from tensorflow.keras.layers import Input, Dense, SimpleRNN, Flatten, LSTM
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import SGD, Adam

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
series = np.sin((0.1 * np.arange(400)) ** 2)

plt.plot(series)

plt.show()
```

```python
T = 10 # 10개의 타임시리즈
D = 1  #  
X = []
Y = []
for t in range(len(series) - T): # 0부터 10개 1부터 10개 ... 0부터 9까지 학습 10을 예측하는 방식으로 계속 누적
    x = series[t: t+T]
    X.append(x)
    y = series[t+T]
    Y.append(y)
```

```python
X = np.array(X).reshape(-1, T) # N x T D 의 형태
Y = np.array(Y)
N = len(X)
print("X shape", X.shape, "Y shape", Y.shape) 
```

## Autoregressive

```python
# Autoregressive RNN model
i = Input(shape=(T, ))
x = Dense(1)(i)
model = Model(i, x)
model.compile(loss='mse', optimizer=Adam(lr=0.01))
            
r = model.fit(X[:-N//2], Y[:-N//2], epochs=80, 
              validation_data=(X[-N//2:], Y[-N//2:]))
```

```python
plt.plot(r.history['loss'], label='loss') # r이 가진 history 영역 중 빼오고 싶은 영역은 loss data... label은 loss로 지정
plt.plot(r.history['val_loss'], label='val_loss')
plt.legend()
```

```python
# One-step forecast using trie targets
outputs = model.predict(X)
print(outputs.shape)
```

```python
predictions = outputs[:,0]
```

```python
plt.plot(Y, label='targets')
plt.plot(predictions, label='predictions')
plt.title('Linear Regression Predictions')
plt.legend()
plt.show()
```

```python
# One-step ( using X[i])
validation_target = Y[-N//2:]
validation_predictions = []
# input of first validation output
i = -N//2
while len(validation_predictions) < len(validation_target):
    #
    p = model.predict(X[i].reshape(1, -1))[0, 0] # 배열형태라서 0콤마0을 통해 단일값으로 가져올 수 있다 1x1 -> scalar
    i += 1
    # update the predictions list
    validation_predictions.append(p)

plt.plot(validation_target, label='forecast_targets')
plt.plot(validation_predictions, label='forecast_predictions')
plt.legend()
```

```python
# Multi-step forecast ( using last_x )
validation_target = Y[-N//2:]
validation_predictions = []


last_x = X[-N//2]

while len(validation_predictions) < len(validation_target):
    #
    p = model.predict(last_x.reshape(1, -1))[0, 0] # 배열형태라서 0콤마0을 통해 단일값으로 가져올 수 있다 1x1 -> scalar
    i += 1
    # update the predictions list
    validation_predictions.append(p)
    
    last_x = np.roll(last_x, -1)
    last_x[-1] = p
    
plt.plot(validation_target, label='forecast_targets')
plt.plot(validation_predictions, label='forecast_predictions')
plt.legend()
# 파란색은 원래 값... 주황색은 전혀 학습이 되지 않은 걸로 보임 . 
```

## Create Model(RNN)

```python
# SimpleRNN을 넣어 커브가 어떻게 바뀌는지 보자
# New try RNN/LSTM model
X.shape
```

```python
# 차수 하나를 더 넣을거얌
X = X.reshape(-1 ,T, 1) # N x T x D 
X.shape 
```

```python
i = Input(shape=(T, D))
x = SimpleRNN(10)(i)
x = Dense(1)(x)
model = Model(i, x)
model.compile(
        loss='mse',
        optimizer=Adam(lr=0.05))
```

```python
# Train the RNN
r = model.fit(X[:-N//2], Y[:-N//2],
         batch_size=32,
         epochs=80,
         validation_data=(X[-N//2:], Y[-N//2:]))
```

```python
plt.plot(r.history['loss'], label='loss') # r이 가진 history 영역 중 빼오고 싶은 영역은 loss data... label은 loss로 지정
plt.plot(r.history['val_loss'], label='val_loss')
plt.legend()
```

```python
outputs = model.predict(X)
print(outputs.shape)
predictions = outputs[:,0]
```

```python
plt.plot(Y, label='targets') 
plt.plot(predictions, label='predictions')
plt.title("many-to-one RNN")
plt.legend()
plt.show() # 일치하진 않지만 파란색 파형을 따라가려한다...
```

```python
# Multi-step forecast 
forecast = []
last_x = X[-N//2]

while len(forecast) < len(Y[-N//2:]):
    new_p = model.predict(last_x.reshape(1, T, 1))[0, 0] 

    # update the predictions list
    forecast.append(new_p)
    
    last_x = np.roll(last_x, -1)
    last_x[-1] = new_p
    
plt.plot(Y[-N//2:], label='targets')
plt.plot(forecast, label='predictions')
plt.title("RNN forecast")
plt.legend()
plt.show()
```
